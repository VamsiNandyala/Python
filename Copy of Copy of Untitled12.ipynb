```python
#write a program print harshad number or not
def dig_sum(num):
  sum=0
  while num:
    var1=num%10
    sum+=var1
    return sum
num=int(input("enter the number:"))

if num%dig_sum(num)==0:
  print("harshad number")
else:
  print("not a harshad number")
```

```python
#dictionary methods
#clear()
car={"carname1":"ford","car name2":"swift","car name3":"honda"}
car.clear()
print(car)
```

```python
#copy()
car={"carname":"ford","carname2":"swift","car name3":"honda"}
car.copy()
print(car)
```

```python
#get()
car={"carname":"ford","carname2":"swift","car name3":"honda"}
x=car.get("carname")
print(x)
```

```python
#items()
car={"carname":"ford","carname2":"swift","carname3":"honda"}
x=car.items()
print(x)
```

```python
#keys()
car={"carname":"ford","carname2":"swift","carname3":"honda"}
x=car.keys()
print(x)
```

```python
#popitem()
car={"carname":"ford","carname2":"swift","carname3":"honda"}
car.popitem()
print(x)
```

```python
#setdefault()
car={"carname":"ford","carname2":"swift","carname3":"honda"}
x=car.setdefault("carname2","vamsi")
print(x)
```

```python
#update()
car={"carname":"ford","carname2":"swift","carname3":"honda"}
car.update({"carname":"benz"})
print(car)
```

```python
#values()
car={"carname":"benz","carname2":"ford","carname3":"honda"}
x=car.values()
print(x)
```
