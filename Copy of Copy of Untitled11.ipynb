```python
# Write a Python program to check if a string starts with a specific prefix.
var=("vamsi","sai")
for i in var:
  if i.startswith("v"):
    print(i)
```

```python
# Write a Python program to check if a string ends with a specific suffix.
var=("vamsi","sai")
for i in var:
  if i.endswith("i"):
    print(i)
```

```python
#Write a Python program that takes a user input and counts the number of vowels (a, e, i, o, u) in the input string.
var=input("enter the alphabets:")
vowels=("a,e,i,o,u")
count=0
for vowel in var:
  if vowel in vowels:
    count=count+1
    print(count)
```

```python
#write a python program that takes a user input and checks if it is a palindrome (read the same backwards and forwards).
var=input("enter the word:")
rev_var=var[::-1]
if var==rev_var:
  print("it is a palindrome")
else:
  print("it is not a palindrome")
```
